## 1-1) datetime 모듈 사용하고 오늘 날짜 date.today()를 통해 받아와서 연애 시작일과 경과한 날짜 출력.

```python
from datetime import date

start = date(2019, 2, 14)
today = date.today()

delta = today - start

print(f'춘향이와 몽룡이의 연애 시작일 : {start.year}년 {start.month}월 {start.day}일')
print(f'연애 시작일로부터 경과한 날짜 : {delta.days}일')
```

## 1-2) 100일, 200일, 500일, 1000일 기념 출력 (timedelta() 함수 사용)

```python
from datetime import date, timedelta

start = date(2019, 2, 14)

days_list = [100, 200, 500, 1000]

print(f'춘향이와 몽룡이의 연애 시작일 : {start.year}년 {start.month}월 {start.day}일')

for i in days_list:
    anniv = start + timedelta(days = i)
    print(f'{i}일 기념일 : {anniv.year}년 {anniv.month}월 {anniv.day}일')
```

## 2-1) 1부터 1000000까지 합을 구하는 시간을 초 단위로 구하여 출력.

```python
import time

def sum1to1000000():
    total = 0
    for i in range(1,1000001):
        total += i
    return total

start = time.time()
result = sum1to1000000()
end = time.time()

print(f'1에서 1000000까지의 합을 구하는 시간 : {end-start:.4f}초')
```

## 2-2) 이 함수를 100번 호출하여 소요된 시간 출력

```python
import time

def sum1to1000000():
    total = 0
    for i in range(1,1000001):
        total += i
    return total

start = time.time()

for i in range(100):
    result = sum1to1000000()

end = time.time()

print(f'1000000까지의 합을 100번 반복해서 구하는 시간 : {end-start:.4f}초')
```

## 3-1) 'Hello Python! 100번 출력하는 데 걸리는 시간 출력

## 3-2) 10000 팩토리얼 구하는 함수 100번 반복해서 구하는 시간

```python
import time

def factorial_10000():
    fact = 1
    for i in range(1,10001):
        fact *= i
    return fact

start = time.time()

for i in range(100):
    result = factorial_10000()

end = time.time()

print(f'10000!을 100번 반복해서 구하는 시간 : {end-start:.4f}초')
```

## 3-3) 1에서 10000까지의 모든 홀수를 세제곱하여 더한 값을 반환하는 함수 100번 반복하는 시간

```python
import time

def sum_even_3():
    total = 0
    for i in range(1,10001,2):
        total += i**3
    return total

start = time.time()

for i in range(100):
    result = sum_even_3()

end = time.time()

print(f'1에서 10000까지 홀수의 세제곱 더하기를 100번 반복하는 시간 : {end-start:.4f}초')
```

## 3-4) 1도에서 360도 사이의 sin 값을 1도 단위로 구하여 그 합을 반환하는 함수 100번 반복하는 시간

```python
import time, math

def sum_sin():
    total = 0
    for i in range(1,361):
        rad = math.radians(i)
        total += math.sin(rad)
    return total

start = time.time()

for i in range(100):
    result = sum_sin()

end = time.time()

print(f'1에서 360도까지 sin의 합을 100번 반복하는 시간 : {end-start:.4f}초')
```

## 4-1) 0부터 1000000까지의 임의의 정수를 연속적으로 10개 출력하는 의사난수를 만들기 위한 계수값과 식

```python
import random

def my_rand():
    return random.randrange(0,1000001)

for i in range(10):
    print(my_rand())
```

## 4-2) 초기 시드 값으로 time 모듈의 time() 함수를 사용하여 다시 코딩. 0에서 1000000까지의 임의의 5개 정수 출력하고 2번 반복.

```python
import random
import time

random.seed(time.time())

for _ in range(2):
    for _ in range(5):
        print(random.randrange(1,1000001))
    print()
```
