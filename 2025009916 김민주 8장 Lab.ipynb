## 8-1 ) datetime 모듈을 사용하여 오늘의 날짜와 시간 출력. 이때 hour 값이 10일 경우 오전 10시, 13이면 오후 1시와 같이 오전/오후 정보 출력.

```python
from datetime import datetime

now = datetime.now()

year = now.year
month = now.month
day = now.day
hour = now.hour
minute = now.minute
second = now.second

if hour <12:
    period = '오전'
    display_hour = hour if hour != 0 else 12
else:
    period = '오후'
    display_hour = hour-12 if hour > 12 else 12

print(f'오늘의 날짜 : {year}년 {month}월 {day}일')
print(f'현재시간 : {period} {display_hour}시 {minute}분 {second}초')
```

## 8-2

### 1) 날짜계산 프로그램 수정하여 오늘 날짜와 함께 2027년 크리스마스까지의 남은 날짜 & 시간 출력.

```python
from datetime import datetime

now = datetime.now()
target = datetime(2027, 12, 25, 0, 0, 0)

delta = target - now

print(f'오늘은 {now.year}년 {now.month}월 {now.day}일입니다.')
print(f'2027년 크리스마스 까지는 {delta.days}일 {delta.seconds // 3600}시간 남았습니다.')
```

### 2) 2036년 1월 1일까지 남은 날짜와 시간 출력

```python
from datetime import datetime

now = datetime.now()
target = datetime(2036,1,1,0,0,0)

delta = target - now

days = delta.days
hours = delta.seconds//3600

print(f'오늘은 {now.year}년 {now.month}월 {now.day}일입니다.')
print(f'2036년 새해 까지는 {days}일 {hours}시간 남았습니다.')
```

### 3) 다가오는 자신의 생일까지 남은 날짜와 시간 출력 (12월30일)

```python
from datetime import datetime

now = datetime.now()
target = datetime(2026,12,30,0,0,0)

delta = target - now

days = delta.days
hours = delta.seconds//3600

print(f'오늘은 {now.year}년 {now.month}월 {now.day}일입니다.')
print(f'2026년 생일 까지는 {days}일 {hours}시간 남았습니다.')
```

## 8-3

### 1) timedelta 클래스 사용하여 오늘 날짜로부터 1000일 후의 날짜 구하기

```python
from datetime import datetime, timedelta

now = datetime.now()

future = now + timedelta(days=1000)

print(f'오늘 날짜 : {now.year}년 {now.month}월 {now.day}일')
print(f'1000일 후 날짜 : {future.year}년 {future.month}월 {future.day}일')
```

### 2) 처음으로 사귄 연도, 월, 일 띄어쓰기로 입력하면 100일 기념일 출력

```python
from datetime import datetime, timedelta

year, month, day = map(int, input('처음으로 사귄 연도와 월, 일을 입력하시오 : ').split())

start = datetime(year, month, day)

anniv = start + timedelta(days=99)

print(f'100일 기념일은 : {anniv.year}년 {anniv.month}월 {anniv.day}일입니다.')
```

## 8-4

### 1) math 모듈과 for-in range를 사용하여 4의 2승부터 10승까지 출력

```python
import math

for i in range(2,11):
    result = math.pow(4,i)
    print(f'4**{i}=  {result}')
```

### 2) 일반 각도 0도에서 180도를 10도 단위로 출력. 이때, 일반각도에 대응하는 라디안 각도 값 함께 출력.

```python
import math

for deg in range(0,181,10):
    rad = math.radians(deg)
    print(f'{deg} degree = {rad:.3f} radian')
```

### 3) math 모듈과 for문 사용하여 일반각도 sin 0도에서부터 180까지의 값을 10도 간격으로 출력.

```python
import math

for i in range(0,181,10):
    rad = math.radians(i)
    sin = math.sin(rad)
    print(f'sin({i})={sin:.2f}')
```

## 8-5

### 1) random 모듈의 randrange() 함수를 이용하여 0에서 100 이하의 정수 중 5의 배수 값을 임의로 3개 선택하여 리스트 형식으로 출력.

```python
import random

nums = []

for _ in range(3):
    nums.append(random.randrange(0,101,5))

print(f'0에서 100 이하의 정수 중에서 5의 배수 : {nums}')
```

### 2) 1에서 10 사이의 정수 중 임의의 정수 3개를 sample() 함수를 이용하여 출력

```python
import random

nums = random.sample(range(1,11), 3)

print(f'1에서 10 사이의 임의의 정수 : {nums}')
```

## 8-6

### 1) random 모듈 이용하여 매주 5개의 랜덤한 로또번호 생성.

```python
import random as rd

print('이번 주의 추천 로또 번호')

for i in range(5):
    lotto_list = list(range(1,46))
    rd.shuffle(lotto_list)
    lotto_list = lotto_list[:6]
    lotto_list.sort()
    print(lotto_list)
```

### 2) 임의의 로또 당첨 번호 + 보너스 번호 출력.

```python
import random as rd

lotto_list = list(range(1,46))
rd.shuffle(lotto_list)
lotto_list = lotto_list[:6]
lotto_list.sort()

bonus = rd.choice([n for n in range(1,46) if n not in lotto_list])

print('이번 주의 당첨 로또 번호 + 보너스')
print(f'{lotto_list} + [{bonus}]')
```
